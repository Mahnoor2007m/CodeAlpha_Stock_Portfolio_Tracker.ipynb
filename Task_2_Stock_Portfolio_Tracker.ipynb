## 📌 Objective

The objective of this project is to build a simple stock portfolio tracker using Python. The program calculates the total investment value based on manually defined stock prices and user input.

---

## 🛠️ Concepts Used

- Dictionary
- Input / Output
- Basic Arithmetic
- Variables

---

## 🎯 Features

- Predefined stock prices
- User enters stock name and quantity
- Calculates investment value
- Displays total investment

```python
# ==========================================
#CodeAlpha Python Programming Internship
#Task 2: Stock Portfolio Tracker
#Developed By: Mahnoor Munawar
# ==========================================

# Predefined stock prices
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 150,
    "MSFT": 200,
    "AMZN": 170
}

print("=" * 40)
print("      STOCK PORTFOLIO TRACKER")
print("=" * 40)

print("\nAvailable Stocks:")

for stock, price in stock_prices.items():
    print(f"{stock} : ${price}")

# User input
stock_name = input("\nEnter Stock Name: ").upper()
quantity = int(input("Enter Quantity: "))

# Check stock
if stock_name in stock_prices:

    price = stock_prices[stock_name]

    total = price * quantity

    print("\n----- Investment Summary -----")
    print("Stock:", stock_name)
    print("Price Per Share: $", price)
    print("Quantity:", quantity)
    print("Total Investment: $", total)

else:
    print("\nInvalid Stock Name!")
```
